```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,roc_auc_score
from sklearn.preprocessing import LabelEncoder
```

```python
table = pd.read_csv('C:/Users/UJJAWAL/OneDrive/Desktop/Assignments/CustomerChurn.csv')
```

```python
table
```

```python
table.head()
```

```python
table.drop(columns=['customerID'],inplace=True)
```

```python
for i in table.select_dtypes(include = ['object']).columns:
    if i != 'Churn':
        table[i] = LabelEncoder().fit_transform(table[i])
```

```python
table['Churn'] = table['Churn'].map({'No':0,'Yes':1})
```

```python
X = table.drop(columns=['Churn'])
Y = table['Churn']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train,Y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred_prob = clf.predict_proba(X_test)[:,1]
```

```python
print(classification_report(Y_test,y_pred))
print('ROC-AUC Score:',roc_auc_score(Y_test,y_pred_prob))
```
